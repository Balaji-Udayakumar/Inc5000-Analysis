## Q2) Data Cleaning
### 1. Handling Missing Values

```python
# Counting the total number of nulls in each column and printing them
print(df.isnull().sum())

# Since the unnamed columns 14 and 15 have lot of null values, we are dropping those columns in the original dataframe
df.dropna(thresh = 24,axis = 1, inplace = True)
df.columns
```

## Q5) A Python function with parameters of data frame, category, value, and a number that specifies the number of rows to display in the result.

```python
# Function to get the top n companies based on revenues for a specific value in the category

def top_n_companies(df, category=None, value=None, n=10):
    
    # Filtering the DataFrame if both category and value are provided
    if category is not None and value is not None:
        filtered_df = df[df[category] == value]
    # Else assigning back the original dataframe
    else:
        filtered_df = df
    
    # Sorting the filtered DataFrame by revenue in descending order and returning the top n rows
    sorted_df = filtered_df.sort_values(by='REVENUE', ascending=False).head(n)
    
    return sorted_df
```

## Q7) Awarding companies in industries with the name "Services" in it based on their revenues

- The top quarter of these companies were given a Trailblazer award
- The companies who didn't make the top quarter, but made the top half were given a Pioneer award
- Finally, a data frame showing the company, its rank, its revenue, and which award it will earn, sorted by revenue in descending order is displayed.

```python
# Identifying companies in service industries
service_companies = df.loc[df['INDUSTRY'].str.contains("Services", case=False, na=False)]
service_companies.head()
```

```python
# Calculate revenue partition for service companies to award them accordingly
revenue_partitions = service_companies['REVENUE'].quantile([0.75, 0.5])
top_quarter_threshold = revenue_partitions[0.75]
top_half_threshold = revenue_partitions[0.5]

trailblazer_award_condition = service_companies['REVENUE'] >= top_quarter_threshold

# Making a copy of the slice and assigning it to trailblazer_service_companies before modifying it
trailblazer_service_companies = service_companies.loc[trailblazer_award_condition].copy()

# Update the 'AWARD' column in the copy
trailblazer_service_companies['AWARD'] = 'Trailblazer'

pioneer_award_condition = (service_companies['REVENUE'] >= top_half_threshold) & (service_companies['REVENUE'] < top_quarter_threshold)

# Making a copy of the slice and assigning it to pioneer_service_companies before modifying it
pioneer_service_companies = service_companies.loc[pioneer_award_condition].copy()

# Updating the 'AWARD' column in the copy
pioneer_service_companies['AWARD'] = 'Pioneer'

# Concatenating the two DataFrames along rows
concatenated_service_companies = pd.concat([trailblazer_service_companies, pioneer_service_companies])

# Creating a data frame with company name, rank, revenue, and award
award_data = concatenated_service_companies[['COMPANY NAME', 'RANK', 'REVENUE', 'AWARD']]
    
# Sorting the data frame by revenue in descending order
award_data = award_data.sort_values(by='REVENUE', ascending=False)

award_data
```

## Q8) Visualization 1

- A bar chart to visualize the distribution of count of companies by top 10 industries

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Counting the number of companies in each industry (top 10 industries)
top_10_industries = df['INDUSTRY'].value_counts().head(10)

# Creating a seaborn bar plot to visualize the distribution of companies by industry
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Setting style
sns.barplot(x=top_10_industries.index, y=top_10_industries.values, palette="muted")
plt.title('Distribution of Companies by Industry (Top 10 Industries)')
plt.xlabel('Industry')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)  # Rotating horizontal axis labels for better readability
plt.show()
```

## Inference 1

From the above bar graph we can see that there were a lot of companies in the Advertising & Marketing sector making a good revenue.

This could be because the U.S. economy was recovering from the global financial crisis of 2008, and as consumer confidence increased, businesses were more willing to invest in advertising and marketing to reach a growing and more stable consumer base.

Also, with the increasing use of smartphones, mobile advertising became a significant source of revenue for companies. Advertisers were quick to adapt their strategies to reach consumers on mobile devices, leading to increased spending in this area.

## Q8) Visualization 2

- A horizontal bar chart to visualize the distribution of top 10 sectors making the highest average revenues.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average revenue for each industry
average_revenue_by_industry = df.groupby('INDUSTRY')['REVENUE'].mean().sort_values(ascending=False)

# Select the top 10 performing sectors by average revenue
top_10_performing_sectors = average_revenue_by_industry.head(10)

# Create a seaborn bar plot to visualize the top-performing sectors
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Setting style
sns.barplot(x=top_10_performing_sectors.values, y=top_10_performing_sectors.index, palette="viridis")
plt.title('Top 10 Performing Sectors by Average Revenue')
plt.xlabel('Average Revenue (in Millions)')
plt.ylabel('Industry')
plt.show()
```

## Inference 2

From the above graph we can see that the companies in the 'Security' sector made the highest average revenue in comparison to the other sectors.

This could be beause from the year 2014 - 2017, there was a heightened awareness of security threats, both physical and digital. Events such as data breaches, cyberattacks, and global security issues contributed to a greater demand for security services.

This surge in demand for cybersecurity services led to a massive boost in the revenues of cybersecurity-focused security companies.
